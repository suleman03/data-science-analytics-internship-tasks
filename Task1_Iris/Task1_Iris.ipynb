# Task 1: Exploring and Visualizing the Iris Dataset

## Objective
The objective of this task is to understand how to load, inspect, summarize, and visualize a dataset using pandas, matplotlib, and seaborn.

The Iris dataset contains flower measurements of three different species.

```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = sns.load_dataset('iris') #load dataset
```

```python
#Display first 5 rows
df.head()
```

```python
#Shape of dataset
df.shape
```

```python
#Column names
df.columns
```

```python
#Dataset information
df.info()
```

```python
df.describe()
```

```python
#This shows relationship between sepal and petal length grouped by species.

plt.figure()
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
#This shows data distribution.

plt.figure()
sns.histplot(df['sepal_length'], kde=True)
plt.title("Distribution of Sepal Length")
plt.show()
```

```python
#This shows spread and outliers.

plt.figure()
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title("Sepal Length by Species")
plt.show()
```


## Conclusion

- The dataset contains 150 samples and 5 columns.
- Petal length and petal width clearly separate species.
- No missing values are present.
- Scatter plots show strong relationships between variables.
- Box plots help identify spread and variation across species.
